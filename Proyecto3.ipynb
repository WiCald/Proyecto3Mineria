```python
import pandas as pd
from tabulate import tabulate

# Cargar el archivo SPSS (.sav)
archivo = "divorcios2023.sav"
df = pd.read_spss(archivo)

# Mostrar información de las columnas (nombres y tipos de datos)
print("\nInformación del DataFrame:")
print(df.info())

# Seleccionar solo las columnas categóricas
columnas_categoricas = df.select_dtypes(include=['category']).columns

# Obtener tablas de frecuencia para cada variable categórica
for col in columnas_categoricas:
    print(f"\nFrecuencia de la variable '{col}':")
    frecuencia = df[col].value_counts().reset_index()
    frecuencia.columns = [col, 'Frecuencia']
    
    # Mostrar solo los 3 primeros y 3 últimos resultados
    print(tabulate(frecuencia.head(5), headers='keys', tablefmt='grid', showindex=False)) 
    print("...")
    print(tabulate(frecuencia.tail(3), headers='keys', tablefmt='grid', showindex=False))  
```

Tabla de Frecuencia Unificada(Todos los años) 

```python
import pandas as pd
from tabulate import tabulate
import glob


archivos = sorted(glob.glob("divorcios20*.sav")) 

# Lista para almacenar los DataFrames
dfs = []


for archivo in archivos:
    anio = archivo[-8:-4]  
    df_temp = pd.read_spss(archivo)
    df_temp["Año"] = int(anio) 
    dfs.append(df_temp)

# Juntar todos los DataFrames
df = pd.concat(dfs, ignore_index=True)

# Identificar variables categóricas (tipo 'category' o 'object')
columnas_categoricas = df.select_dtypes(include=['category', 'object']).columns

# Obtener y mostrar tablas de frecuencia para variables categóricas
for col in columnas_categoricas:
    frecuencia = df[col].value_counts().reset_index()
    frecuencia.columns = [col, 'Frecuencia']

    print(f"\nFrecuencia de la variable '{col}':")
    # Usamos 'grid' para que sea más claro y bonito para copiar
    print(tabulate(frecuencia.head(12), headers='keys', tablefmt='grid', showindex=False))


```

## Analisis de correlacion Variables

```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def clean_age_data(age_series):
    #Limpieza de datos de edad
    age_str = age_series.astype(str)
    age_str = age_str.replace(['Ignorado', 'N/A', 'NaN', 'nan', 'NULL', 'null', '?', '-'], np.nan)
    age_numeric = pd.to_numeric(age_str, errors='coerce')
    age_filtered = age_numeric[(age_numeric >= 15) & (age_numeric <= 100)]
    return age_filtered

def correlaciones(df):
    # Limpieza inicial
    df_clean = df.copy()
    df_clean['EDADHOM_clean'] = clean_age_data(df['EDADHOM'])
    df_clean['EDADMUJ_clean'] = clean_age_data(df['EDADMUJ'])
    df_clean['DIF_EDAD'] = df_clean['EDADHOM_clean'] - df_clean['EDADMUJ_clean']
    
    #EDADES
    datos_edad = df_clean[['EDADHOM_clean', 'EDADMUJ_clean']].dropna()
    if len(datos_edad) >= 2:
        x = datos_edad['EDADHOM_clean'].values
        y = datos_edad['EDADMUJ_clean'].values
        
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        
        plt.figure(figsize=(10,6))
        plt.scatter(x, y, alpha=0.5, label='Datos')
        plt.plot(x, slope*x + intercept, 'r-', 
                 label=f'r = {r_value:.2f}, p = {p_value:.3f}\ny = {slope:.2f}x + {intercept:.2f}')
        plt.xlabel('Edad del Hombre')
        plt.ylabel('Edad de la Mujer')
        plt.title('Correlación entre Edades al Divorcio')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        print("\n=== CORRELACIÓN ENTRE EDADES ===")
        print(f"Correlación Pearson: r = {r_value:.2f} (p = {p_value:.4f})")
        print(f"Pares válidos: {len(x)}")
    
    #EDUCACION Y DIF EDAD
    if 'ESCHOM' in df_clean.columns:
        valores_educ = df_clean['ESCHOM'].dropna().unique()
        print("\nValores únicos en ESCHOM:", valores_educ)
        
        # Definir orden basado en valores existentes
        educ_order = ['Ninguno', 'Primaria', 'Básico', 'Diversificado', 'Universitario']
        educ_order = [x for x in educ_order if x in valores_educ]
        
        if len(educ_order) > 1:  #Al menos 2
            try:
                df_clean['ESCOLARIDAD_NUM'] = (
                    df_clean['ESCHOM']
                    .astype('category')
                    .cat.set_categories(educ_order, ordered=True)
                    .cat.codes
                )
                
                educ_data = df_clean[['ESCOLARIDAD_NUM', 'DIF_EDAD']].dropna()
                
                if len(educ_data) > 1:
                    r_educ, p_educ = stats.pearsonr(
                        educ_data['ESCOLARIDAD_NUM'], 
                        educ_data['DIF_EDAD']
                    )
                    
                    plt.figure(figsize=(10, 6))
                    plt.scatter(educ_data['ESCOLARIDAD_NUM'], educ_data['DIF_EDAD'], 
                               alpha=0.5, color='green')
                    plt.xticks(range(len(educ_order)), educ_order, rotation=45)
                    plt.xlabel('Nivel Educativo del Hombre')
                    plt.ylabel('Diferencia de Edad (Hombre - Mujer)')
                    plt.title(f'Correlación: r = {r_educ:.2f} (p = {p_educ:.3f})')
                    plt.grid(True)
                    plt.show()
                    
                    print("\n=== CORRELACIÓN EDUCACIÓN vs DIFERENCIA DE EDAD ===")
                    print(f"Correlación Pearson: r = {r_educ:.2f} (p = {p_educ:.4f})")
                    print("Interpretación: Valores positivos indican mayor diferencia de edad con mayor educación")
            except Exception as e:
                print(f"\nError al procesar educación: {str(e)}")
    
    #MES OCURRENCIA Y REGISTRO
    if all(col in df_clean.columns for col in ['MESOCU', 'MESREG']):
        meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio',
                'Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
        
        meses_existentes = list(
            set(df_clean['MESOCU'].dropna().unique()) | 
            set(df_clean['MESREG'].dropna().unique())
        )
        meses = [m for m in meses if m in meses_existentes]
        
        if len(meses) > 1:
            try:
                df_clean['MESOCU_NUM'] = (
                    df_clean['MESOCU']
                    .astype('category')
                    .cat.set_categories(meses, ordered=True)
                    .cat.codes
                )
                df_clean['MESREG_NUM'] = (
                    df_clean['MESREG']
                    .astype('category')
                    .cat.set_categories(meses, ordered=True)
                    .cat.codes
                )
                
                meses_data = df_clean[['MESOCU_NUM', 'MESREG_NUM']].dropna()
                
                if len(meses_data) > 1:
                    r_meses, p_meses = stats.pearsonr(
                        meses_data['MESOCU_NUM'], 
                        meses_data['MESREG_NUM']
                    )
                    
                    plt.figure(figsize=(10, 6))
                    plt.scatter(meses_data['MESOCU_NUM'], meses_data['MESREG_NUM'], 
                               alpha=0.5, color='purple')
                    plt.xlabel('Mes de Ocurrencia (numérico)')
                    plt.ylabel('Mes de Registro (numérico)')
                    plt.title(f'Correlación: r = {r_meses:.2f} (p = {p_meses:.3f})')
                    plt.grid(True)
                    plt.show()
                    
                    print("\n=== CORRELACIÓN MES OCURRENCIA Y REGISTRO ===")
                    print(f"Correlación Pearson: r = {r_meses:.2f} (p = {p_meses:.4f})")
                    print("Interpretación: Valores cercanos a 1 indican poca diferencia entre ocurrencia y registro")
            except Exception as e:
                print(f"\nError al procesar meses: {str(e)}")
    
    #DEPARTAMENTO Y EDAD
    if 'DEPREG' in df_clean.columns:
        top_deptos = df_clean['DEPREG'].value_counts().nlargest(5).index
        df_top = df_clean[df_clean['DEPREG'].isin(top_deptos)]
        
        if len(top_deptos) > 1:
            grupos = [df_top[df_top['DEPREG'] == depto]['EDADHOM_clean'].dropna() 
                     for depto in top_deptos]
            
            if all(len(g) > 1 for g in grupos):
                f_val, p_val = stats.f_oneway(*grupos)
                
                plt.figure(figsize=(12, 6))
                df_top.boxplot(column='EDADHOM_clean', by='DEPREG', grid=True)
                plt.title('Edad del Hombre por Departamento')
                plt.suptitle('')
                plt.xlabel('Departamento')
                plt.ylabel('Edad')
                plt.xticks(rotation=45)
                plt.grid(True)
                plt.show()
                
                print("\n=== CORRELACIÓN GEOGRÁFICA ===")
                print(f"ANOVA: F = {f_val:.2f}, p = {p_val:.4f}")
                print("Diferencias significativas entre departamentos" if p_val < 0.05 
                      else "No hay diferencias significativas entre departamentos")

archivos = sorted(glob.glob("divorcios20*.sav"))
dfs = []
for archivo in archivos:
    year = archivo[-8:-4]
    df_temp = pd.read_spss(archivo)
    df_temp["AÑOREG"] = int(year)
    dfs.append(df_temp)
df = pd.concat(dfs, ignore_index=True)

# Ejecutar análisis
correlaciones(df)
```
