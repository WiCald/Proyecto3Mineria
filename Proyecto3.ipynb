Año 2023

```python
import pandas as pd
from tabulate import tabulate

# Cargar el archivo SPSS (.sav)
archivo = "divorcios2023.sav"
df = pd.read_spss(archivo)

# Mostrar información de las columnas (nombres y tipos de datos)
print("\nInformación del DataFrame:")
print(df.info())

# Seleccionar solo las columnas categóricas
columnas_categoricas = df.select_dtypes(include=['category']).columns

# Obtener tablas de frecuencia para cada variable categórica
for col in columnas_categoricas:
    print(f"\nFrecuencia de la variable '{col}':")
    frecuencia = df[col].value_counts().reset_index()
    frecuencia.columns = [col, 'Frecuencia']
    
    # Mostrar solo los 3 primeros y 3 últimos resultados
    print(tabulate(frecuencia.head(5), headers='keys', tablefmt='grid', showindex=False)) 
    print("...")
    print(tabulate(frecuencia.tail(3), headers='keys', tablefmt='grid', showindex=False))  
```

Tabla de Frecuencia Unificada(Todos los años) 

```python
import pandas as pd
from tabulate import tabulate
import glob


archivos = sorted(glob.glob("divorcios20*.sav")) 

# Lista para almacenar los DataFrames
dfs = []


for archivo in archivos:
    anio = archivo[-8:-4]  
    df_temp = pd.read_spss(archivo)
    df_temp["Año"] = int(anio) 
    dfs.append(df_temp)

# Juntar todos los DataFrames
df = pd.concat(dfs, ignore_index=True)

# Identificar variables categóricas (tipo 'category' o 'object')
columnas_categoricas = df.select_dtypes(include=['category', 'object']).columns

# Obtener y mostrar tablas de frecuencia para variables categóricas
for col in columnas_categoricas:
    frecuencia = df[col].value_counts().reset_index()
    frecuencia.columns = [col, 'Frecuencia']

    print(f"\nFrecuencia de la variable '{col}':")
    # Usamos 'grid' para que sea más claro y bonito para copiar
    print(tabulate(frecuencia.head(12), headers='keys', tablefmt='grid', showindex=False))


```
